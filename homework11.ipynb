Read about the [pandas.pivot_table](https://pandas.pydata.org/docs/reference/api/pandas.pivot_table.html) function and analyze how it works in the provided examples.

Next, using the data provided below in the `df_employees` DataFrame, use the `pivot_table` function to create a summary DataFrame. The index should consist of the various types of employment, i.e., the unique values of the 'Type' column. The columns should contain the sum of salaries, the average salary, and the number of employees separately for each department, i.e., for each unique value in the 'Department' column.

```python
import pandas as pd

df_employees = pd.DataFrame({'First Name': ['Tom', 'Adam', 'Jane', 'Alice', 'Robert', 'Bob', 'John', 'Frank', 'Eva',
                                            'John', 'Lois'],
                   'Last Name': ['Brown', 'Green', 'Thompson', 'Smith', 'Newman', 'Parker', 'Williams', 'Taylor', 'Evans',
                                 'Lewis', 'White'],
                   'Type': ['Full-time Employee', 'Intern', 'Full-time Employee', 'Part-time Employee',
                            'Full-time Employee', 'Intern', 'Intern', 'Part-time Employee', 'Part-time Employee',
                            'Full-time Employee', 'Part-time Employee'],
                   'Department': ['Administration', 'Technical', 'Administration', 'Technical', 'Management',
                                  'Administration', 'Management', 'Administration', 'Management', 'Technical',
                                  'Management'],
                   'YearsOfExperience': [4, 3, 5, 7, 6, 1, 2, 2, 4, 5, 2],
                   'Salary': [15000, 6000, 9000, 10000, 20000, 3000, 4000, 6000, 12000, 8000, 5500]})

# df_employees
```

```python
pd.pivot_table(df_employees,index=['Department','Type'], values='Salary',aggfunc=['sum','mean','count'])
```
